Логистическая регрессия. Новиков Дмитрий. Вариант "Финики 2".

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Finiki1.csv", sep=',')
data.drop(columns = ['Unnamed: 0'], inplace = True)
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
data['AREA'] = data.apply(lambda x: float(x['AREA']), axis = 1)
data['MAJOR_AXIS'] = data.apply(lambda x: float(x['MAJOR_AXIS']), axis = 1)
data['PERIMETER'] = data.apply(lambda x: float(x['PERIMETER']), axis = 1)
data['ECCENTRICITY'] = data.apply(lambda x: float(x['ECCENTRICITY']), axis = 1)
data
```

## Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression

X_RL = data[['AREA','PERIMETER', 'MAJOR_AXIS', 'ECCENTRICITY']]
y_RL = data['Class']
X_train, X_test, y_train, y_test = train_test_split(X_RL, y_RL, random_state = 0)
clf = LogisticRegression().fit(X_train, y_train)
```

Строчка clf = LogisticRegression().fit(X_train, y_train) подбирает модель - то есть по значениям X_train(значения параметров), y_train(единицы и нули) строит оценку максимума правдоподобия для коэффициентов, то есть находит коэффициенты, которые будет использовать в дальнейшем

```python
print('Accuracy of Logistic regression classifier on training set: %.30f'%clf.score(X_train, y_train))
print('Accuracy of Logistic regression classifier on test set: {:.30f}'.format(clf.score(X_test, y_test)))
```

```python
X_RL1 = data[['ECCENTRICITY','AREA']]
y_RL1 = data['Class']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_RL1, y_RL1,random_state = 0)
clf1 = LogisticRegression().fit(X_train1, y_train1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 4)
model.fit(X_train,y_train)
```
